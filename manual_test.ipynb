test area and perimeter of circle

```python
from circle import Circle

c1 = Circle(1, 1, 1)

print(c1.area)
print(c1.perimeter)
```

```python
c2 = Circle('one', 2, 3)

try:
    print(c2.__repr__)
except TypeError as err:
    raise err ('must be a numeric value')
```

```python
from rectangle import Rectangle

r1 = Rectangle(2, 2, 2, 2)

print(r1.area)
print(r1.perimeter)
```
